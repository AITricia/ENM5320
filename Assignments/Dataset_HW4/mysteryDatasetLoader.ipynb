```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

The following will load a pickle file containing 6 different solutions to a mystery nonlinear wave equation as a list of time vs space pts. Plots are generated providing the initial and final conditions.

```python
# %% Load solutions from pickle file and plot
with open('solutions.pkl', 'rb') as f:
    solutions = pickle.load(f)
# Plot the solutions at initial time
plt.figure(figsize=(10, 6))
for k, u in enumerate(solutions, 1):
    plt.plot(u[0, :], label=f'k = {k}')
plt.xlabel('x')
plt.ylabel('u')
plt.title('Initial condition for Different k')
plt.legend()
plt.show()
# Plot the solutions at final time
plt.figure(figsize=(10, 6))
for k, u in enumerate(solutions, 1):
    plt.plot(u[500, :], label=f'k = {k}')
plt.xlabel('x')
plt.ylabel('u')
plt.title('Final solution for Different k')
plt.legend()
plt.show()
```
